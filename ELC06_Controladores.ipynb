\label{H_95DCEBB1}
\matlabheading{Control Proporcional}

\begin{par}
\begin{flushleft}
El controlador proporcional o \textbf{controlador P} es aquel que se obtiene al multiplicar al sistema por una ganacia $K_p$. Al analizar el diagramas de bloque asociado, es importante destacar las similitudes y las diferencias con este tipo de controlador y el diagrama que utiliza internamente la función \textbf{rlocus}
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
\includegraphics[width=\maxwidth{96.53788258906172em}]{image_0}
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Si bien el control proporcional y el root locus son similares, no son idénticos. Esto se debe a que \textbf{rlocus} asume que la retroalimentación está en la trayectoría de retroalimentación y no en la trayectoría directa. Para evaluar que implica esto, es necesario hacer la reducción de bloques en cada caso.
\end{flushleft}
\end{par}


\vspace{1em}
\begin{par}
\begin{flushleft}
En el caso del \textbf{Control proporcional} la reducción de bloques queda
\end{flushleft}
\end{par}

\begin{par}
$$\frac{Y(s)}{U(s)}~\Rightarrow ~\frac{K_p \cdot Y(s)}{U(s)}~\Rightarrow ~\frac{\frac{K_p \cdot Y(s)}{U(s)}}{1+\frac{K_p \cdot Y(s)}{U(s)}}~\Rightarrow ~\frac{K_p \cdot Y(s)}{U(s)+K_p \cdot Y(s)}$$
\end{par}

\begin{par}
\begin{flushleft}
Y en el caso del \textbf{Root Locus} la reducción de bloques queda
\end{flushleft}
\end{par}

\begin{par}
$$\frac{Y(s)}{U(s)}~\Rightarrow ~\frac{\frac{Y(s)}{U(s)}}{1+K_p \cdot \frac{Y(s)}{U(s)}}~\Rightarrow ~\frac{Y(s)}{U(s)+K_p \cdot Y(s)}$$
\end{par}

\begin{par}
\begin{flushleft}
Se puede observar como los \textbf{denominadores }son iguales mientras que los numeradores no lo son, surje entonces la pregunta ¿es esta diferencia relevante? La respuesta es que no, ya que el \textbf{rlocus} sólo muestra \textbf{polos} y \textbf{ceros}. Como el denominador es el mismo, ambos sistemas tienen la misma ecuación característica, y por ende, los mismos polos. Por otro lado, como $K_p$ es una constante, las posiciones de los ceros tampoco se ven alteradas.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Esto no quiere decir que ambos sistemas sean \textit{equivalentes} ya que la \textbf{respuesta} del sistema en cada caso va a ser diferente, pero los ceros y polos se mantendrán como estaban originalmente.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Entonces el método de \textbf{rlocus} nos permite evaluar exactamente donde se situarán los polos en lazo cerrado al utilizar la función en lazo abierto \textit{\textbf{como si}} estuvieramos usando un controlador Proporcional.
\end{flushleft}
\end{par}

\label{H_DA2AE5E1}
\matlabheadingtwo{Efectos del controlador proporcional}

\begin{par}
\begin{flushleft}
Como se mostró en la sección anterior, un controlador P \textbf{no} agrega polos ni ceros, sin embargo, \textbf{si} modifica la posición de los ya existentes.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Aumentar el valor de $K_p$ tiene los siguientes efectos en las métricas de rendimiento:
\end{flushleft}
\end{par}

\begin{itemize}
\setlength{\itemsep}{-1ex}
   \item{\begin{flushleft} \textbf{Rise Time}: Disminuye \end{flushleft}}
   \item{\begin{flushleft} \textbf{Overshoot}: Aumenta \end{flushleft}}
   \item{\begin{flushleft} \textbf{Settlling time:} Casi sin efecto \end{flushleft}}
   \item{\begin{flushleft} \textbf{Stabilidad Relativa}: Empeora \end{flushleft}}
   \item{\begin{flushleft} \textbf{Steady-State Error}: Disminuye \end{flushleft}}
\end{itemize}

\begin{par}
\begin{flushleft}
Para llegar a esas conclusiones basta con analizar el efecto del controlador en sí. Al aumentar $K_p$ lo que se hace es aumentar el \textbf{error} con respecto a la entrada de manera artificial, esto hace que el sistema reaccione como si el error fuera más grande de lo que realmente es y de esta manera toma cambios más abruptos, provocando un \textbf{pico más pronto} y \textbf{aumentando la amplitud del pico. }Como el error luego disminuye drásticamente, también lo hace la señal de salida del controlador y por lo tanto el tiempo de asentamiento apenas se ve modificado.
\end{flushleft}
\end{par}

\matlabheadingthree{\textbf{Siempre inestable}}

\begin{par}
\begin{flushleft}
Esto ocurre cuando una de las ramas o (loci) del root locus está totalmente a la derecha del eje imaginario.
\end{flushleft}
\end{par}

\begin{matlabcode}
ceros = [-1.5];
polos = [-1 2 -0.5+2j -0.5-2j];
[num, den] = zp2tf(ceros, polos, [1]);
sistema = tf(num, den);
rlocus(sistema)
title("Sistema inestable para todo valor de k_p")
\end{matlabcode}
\begin{center}
\includegraphics[width=\maxwidth{91.82137481184145em}]{figure_1.png}
\end{center}

\begin{par}
\begin{flushleft}
También es posible que todas las ramas tengan una parte en el lado izquierdo del eje imaginario pero que no exista un valor de $K_p$ que haga que todos los polos se ubiquen \textbf{simultaneamente} a la derecha del eje imaginario. En el ejemplo siguiente, para que las ramas asociadas a los polos complejos se mantengan a la izquierda del eje imaginario, $K_p$ tiene que ser \textbf{menor }que \textasciitilde{}$5.2$ pero al mismo tiempo, para que la rama asociada con el polo $2$ entre en el semiplano izquierdo, $K_p$ tiene que ser \textbf{mayor }que \textasciitilde{}$9.70$. De esta manera se concluye que el sistema es siempre inestable ya que $K_p$ no puede ser menor que $5.2$ y simultaneamente mayor que $9.70$
\end{flushleft}
\end{par}

\begin{matlabcode}
ceros = [-2];
polos = [-1 2 -0.2+3j -0.2-3j];
[num, den] = zp2tf(ceros, polos, [1]);
sistema = tf(num, den);
rlocus(sistema)
title("Sistema inestable para todo valor de k_p")
\end{matlabcode}

\label{H_E2540CC0}
\matlabheadingthree{\textbf{Estabilidad dependiende de K}}

\begin{par}
\begin{flushleft}
Esto ocurre cuando una o más ramas tienen algunas de sus secciones a la derecha del eje imaginario, esto quiere decir que para un $K_p$ suficientemente grande, el sistema se vuelve inestable, pero para valores menores a ese $K_p$ crítico, el sistema es estable. A simple vista, siempre que haya una única rama que cruza al semiplano derecho (o un par de ramas simétricas), va a existir un $K_p$ para el cual el sistema va a ser estable, si hay varias ramas no simétricas que pasan al semiplano derecho, es necesario evaluar el $K_p$ crítico para cada una, como se hizo en el caso anterior.
\end{flushleft}
\end{par}

\begin{matlabcode}
ceros = transpose([-2]);
polos = [-3 -0.5 -2.5+2j -2.5-2j];
[num, den] = zp2tf(ceros, polos, [1]);
sistema = tf(num, den);
rlocus(sistema)
title("Sistema estable para algunos valores de k_p")

\matlabheadingtwo{Error en estado estacionario}

\begin{par}
\begin{flushleft}
Una vez verificado que el sistema es estable (para un $K_p$ determinado), el error en estado estable se puede calcular, pero, éste no depende sólo del valor de $K_p$ sino del tipo de sistema y de la entrada, si el sistema tenía error infinito o cero antes de aplicar el controlador, ese error se va a mantener (ya que el controlador P no agrega polos ni ceros). Si el error era distinto de 0 y finito, entonces se puede calcular con la fórmula del error en estado estable (luego de haber verificado que el sistema es estable), haciendo uso del teorema del valor final. Como todos los términos de $s$ tienden a $0$, el error queda definido únicamente por los términos independientes del numerados ($a_0$) y del denominador ($b_0$).
\end{flushleft}
\end{par}

\begin{par}
\begin{center}
Escalón $\frac{1}{s}$: $E_{ss} =\frac{1}{1+\frac{K_p \cdot a_0 }{b_0 }}~\Rightarrow ~\frac{1}{1+K_p }$
\end{center}
\end{par}

\begin{par}
\begin{center}
Rampa $\frac{1}{s^2 }$: $E_{ss} =\frac{1}{K_p \cdot \frac{a_0 }{b_0 }}~\Rightarrow ~\frac{1}{K_v }$
\end{center}
\end{par}

\begin{par}
\begin{center}
Parábola $\frac{1}{s^3 }$: $E_{ss} =\frac{1}{K_p \cdot \frac{a_0 }{b_0 }}~\Rightarrow ~\frac{1}{K_a }$
\end{center}
\end{par}

\begin{par}
\begin{flushleft}
A los valores $K_p$, $K_v {}$ y $K_a$ se los denomina constante de error de \textbf{posición}, \textbf{velocidad} y \textbf{aceleración} respectivamente.
\end{flushleft}
\end{par}


\vspace{1em}
\begin{par}
\begin{flushleft}
\textbf{¡Advertencia con la notación!}
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Al término $\frac{K_p \cdot a_0 }{b_0 }$ se lo suele abreviar como $K_p$ cuando se trata de una entrada escalón, pero esa $p$ se refiere \textbf{posición} y no a \textbf{proporcional} ya que esta nueva constante $K_p$ \textbf{no }necesariamente coincide con la ganancia $K_p$ del controlador. El único caso donde coincidirían sería cuando $a_0 =1$ y $b_0 =1$. Si bien la expresión no está definida para el valor $b_0 =0$ esto no es un problema ya que en ese caso el sistema sería de un tipo superior y por ende el error en estado estacionario sería 0.
\end{flushleft}
\end{par}


\vspace{1em}
\begin{par}
\begin{flushleft}
\textbf{Error y tipo de sistema}
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
El error no sólo se define por la entrada sino también por el \textbf{tipo de sistema}, el tipo de sistema hace referencia a la cantidad de polos en el origen, otra formas equivalentes de determinar el tipo de sistema son:
\end{flushleft}
\end{par}

\begin{itemize}
\setlength{\itemsep}{-1ex}
   \item{\begin{flushleft} Cantidad de integradores puros \end{flushleft}}
   \item{\begin{flushleft} Cantidad de polos en el origen de coordenadas \end{flushleft}}
   \item{\begin{flushleft} Grado del factor común $s$ del denominador \end{flushleft}}
   \item{\begin{flushleft} Grado de $s$ del término con menor grado del denominador \end{flushleft}}
\end{itemize}

\matlabheading{Diseño de Controladores}

\begin{par}
\begin{flushleft}
A la hora de diseñar un controlador es necesario en primera medida determinar que modificaciones hacen falta en el sistema para poder alcanzar los requisitos de diseño.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Los controladores vistos permiten hacer lo siguiente:
\end{flushleft}
\end{par}

\begin{itemize}
\setlength{\itemsep}{-1ex}
   \item{\begin{flushleft} \textbf{Control Proporcional}: Mueve los polos en lazo cerrado \end{flushleft}}
   \item{\begin{flushleft} \textbf{Control Proporcional Derivativo}: Agrega un Cero Real y mueve los polos en lazo cerrado \end{flushleft}}
   \item{\begin{flushleft} \textbf{Control Proporcional Integral}: Agrega un Polo en el origen y mueve los polos en lazo cerrado \end{flushleft}}
   \item{\begin{flushleft} \textbf{Control Proporcional Integral Derivativo}: Agrega un Polo en el origen y dos ceros (reales o complejos) y mueve los polos en lazo cerrado \end{flushleft}}
\end{itemize}

\begin{par}
\begin{flushleft}
Debido a estas características intrínsecas de los controladores, es útil expresar los requerimientos de diseño en función de los ceros y polos cuando sea posible. Algunos requisitos puntuales pueden no poder expresarse de esta forma pero otros como el overshoot si pueden.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Otra consideración importante es que los controladores \textbf{no} modifican la planta que controlan, esto quiere decir que todos los ceros y polos existentes no pueden moverse, no obstante, aquellos que introduce el controlador si. Para ver esto de manera gráfica puede utilizarse el \texttt{\textbf{Control System Designer}} y, si bien es posible hacerlo mediante la interfaz gráfica como se mostró al hacer el Root Locus. Como en este caso la planta tiene una función de transferencia conocida, es posible abrir la aplicación con la planta ya inicializada mediante el comando \texttt{\textbf{controlSystemDesigner}} y agregará automaticamente la salida ante una entrada escalón y el root locus.
\end{flushleft}
\end{par}

\begin{matlabcode}
sistema = tf([1 3], [1 5 9 5]);
controlSystemDesigner(sistema);
\end{matlabcode}
