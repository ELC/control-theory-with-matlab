\label{H_12CF5A3C}
\matlabheadingtwo{Criterio General de Nyquist}

\begin{par}
\begin{flushleft}
El criterio general de estabilidad de Nyquist puede representarse con la siguiente ecuación
\end{flushleft}
\end{par}

\begin{par}
$$Z=N+P$$
\end{par}

\begin{par}
\begin{flushleft}
$Z$: Cantidad de \textbf{polos del sistema }a la derecha del eje imaginario
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
$N$: Cantidad de rodeos al punto $(-1,0)$, si es en sentido horario es positivo y si es antihorario negativo
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
$P$: Cantidad de \textbf{polos en lazo abierto} a la derecha del eje imaginario
\end{flushleft}
\end{par}


\vspace{1em}
\label{H_A04D7BE4}
\matlabheadingthree{\textbf{Interpretación}}

\begin{par}
\begin{flushleft}
Si el sistema es \textbf{estable}, entonces $Z=0$
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Si los rodeos al punto $(-1,0)$ ocurren en sentido \textbf{horario}, el sistema es inestable, ya que $P$ no puede ser negativo.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Si el sistema es de \textbf{fase mínima}, entonces $P=0$. Por lo tanto, para que el sistema sea estable, \textbf{no tiene que haber rodeos} al punto $(-1,0)$, es decir $N=0$
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Si el sistema es de \textbf{fase no mínima}, pero no tiene polos en lazo abierto a la derecha del eje imaginario, entonces $P=0$. Por lo tanto, para que el sistema sea estable, \textbf{no tiene que haber rodeos} al punto $(-1,0)$, es decir $N=0$
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
Si el sistema es de \textbf{fase no mínima }con polos a la derecha del eje imaginario, entonces $P\not= 0$. Por lo tanto, para que el sistema sea estable se tienen que dar exactamente $P$ rodeos al punto $(-1,0)$ en sentido \textbf{antihorario}.
\end{flushleft}
\end{par}

\matlabheadingtwo{Ejemplo con Fase Mínima}

\begin{par}
\begin{flushleft}
Al ser de fase mínima, $P=0$ y para que el sistema sea estable se requiere que no haya encierros las punto $(-1,0)$. 
\end{flushleft}
\end{par}

\begin{matlabcode}
set(0, 'defaultFigurePosition', [0 0 16 10])

sistema = tf([1], [1 4 6 5 2]);

figure;
subplot(2, 2, [1 2])
rlocus(sistema)

subplot(2, 2, 3)
nyquist(sistema, 3.9375*sistema, 5*sistema)
legend("K = 1", "K = 3.9375", "K = 5")

subplot(2, 2, 4)
lazo_cerrado_1 = feedback(sistema, 1);
lazo_cerrado_2 = feedback(3.9375*sistema, 1);
lazo_cerrado_3 = feedback(5*sistema, 1);
tiempo = 0:0.02:20;
step(lazo_cerrado_1, lazo_cerrado_2, lazo_cerrado_3, tiempo )
ylim([0 1.5])
xlim([0 20])
legend("K = 1", "K = 3.9375", "K = 5")

\matlabheadingtwo{Ejemplo Sistema Fase No Mínima con $P=0$}

\begin{par}
\begin{flushleft}
Al tener un cero a la derecha del eje imaginario, este sistema es de fase no mínima, pero al no tener polos en lazo abierto a la derecha del eje imaginario, $P=0$. Por lo tanto para que el sistema sea estable es necesario que $N=0$, esto quiere decir que no debe haber encierros al punto $(-1,0)$, ni en sentido horario ni antihorario.
\end{flushleft}
\end{par}

\begin{par}
\begin{flushleft}
En este caso $P=1$, por lo tanto, para que el sistema sea estable es necesario que $N=-1$, es decir, que el punto $(-1,0)$ tenga un encierro en sentido antihorario. En el Root Locus puede verse como hay dos potenciales valores críticos de $K$, uno de ellos se encuentra en $\frac{123}{8}$ y el otro en $\approx 52.5134$.
